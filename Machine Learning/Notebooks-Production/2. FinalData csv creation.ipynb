```python
merge_crop = pd.read_csv('../Datasets/MergeFileCrop.csv')
reco_fert = merge_fert
```

```python
import random
import pandas as pd

temp = pd.DataFrame(columns=['N', 'P', 'K'])

for i in range(0, merge_crop.shape[0]):
    crop = merge_crop.label.iloc[i]

    # Check if there are rows matching the condition
    matching_rows = reco_fert[reco_fert['Crop'] == crop]

    if not matching_rows.empty:
        # Access the first matching row
        N = matching_rows["N"].iloc[0] + random.randint(-20, 20)
        P = matching_rows["P"].iloc[0] + random.randint(-5, 20)
        K = matching_rows["K"].iloc[0] + random.randint(-5, 5)

        d = {"N": N, "P": P, "K": K}
        temp = temp.append(d, ignore_index=True)
    else:
        # Handle the case where there are no matching rows for the crop
        print(f"No recommendations found for crop: {crop}")

# Print the resulting DataFrame
print(temp)
```
